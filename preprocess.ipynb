```python
def dig(path, ftype):
    """
    returns list of all files of the specified extension in every recursive subdirectory in a given parent dir
    """
    paths = []
    for dirpath, _, files in os.walk(path):
        paths += [(dirpath + "/" + f) for f in files if f.endswith(ftype)]
    return paths
```

### MXM xref MillionSongs 

```python
songs = {}
for x in range(len(files)):
    with tables.open_file(files[x], 'r') as f:
        name = h5.get_artist_name(f).decode('UTF-8')
        title = h5.get_title(f).decode('UTF-8')
        track_id = h5.get_track_id(f).decode('UTF-8')
        songs[track_id] = (name, title)
print("done")
```

```python
with open(MXM_TRAIN, 'r') as filein, open('mxm_new_train.csv', 'w') as newfile:
    for line in filein:
        cols = line.split(',')
        new_delim = ';'.join(cols[:2]) + ';' + ','.join(cols[2:])
        newfile.write(new_delim)
```

```python
mxm = pd.read_csv("mxm_new_train.csv", delimiter=";", header=None, names=['track_ID', 'ID', 'word_counts'])

def convert_to_dict(word_count_string):
    pairs = word_count_string.split(',')
    word_count_dict = {int(pair.split(':')[0]): int(pair.split(':')[1]) for pair in pairs}
    return word_count_dict

mxm['word_counts'] = mxm['word_counts'].apply(convert_to_dict)
mxm
```

```python
mxm["Artist"] = None
mxm["Title"] = None
mxm
```

```python
for idx, row in mxm.iterrows():
    track_id = row['track_ID']
    if track_id in songs:
        mxm.at[idx, 'Artist'], mxm.at[idx, 'Title'] = songs[track_id]
mxm
```

```python
mxm.dropna(subset=['Artist'], inplace=True)
mxm.drop(columns=["track_ID", "ID"], inplace=True)
mxm
```

```python
mxm.to_csv("mxm_train.csv", ";")
```
